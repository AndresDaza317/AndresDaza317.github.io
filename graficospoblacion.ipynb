```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("poblacion.csv")
df

#graficas

#Datos de ejemplo
x = df["Date"]
y = df["COL"]
z = df["VEN"]
z1 = df["PER"]

plt.plot(x,y, label="COL")
plt.plot(x,z, label="VEN")
plt.plot(x,z1, label="PER")
plt.title("Gr√°fico de poblacion Colombia")
plt.legend()
plt.xlabel("date")
plt.ylabel("COL")
plt.show()
```
